```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_excel('Book1.xlsx')
```

```python
X = data[['Outlook', 'Temp', 'Humidity', 'Windy']]  # Replace with actual feature names
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
dbscan = DBSCAN(eps=0.5, min_samples=5)  # You can adjust the parameters as needed
```

```python
dbscan.fit(X_scaled)
```

```python
labels = dbscan.labels_
```

```python
plt.scatter(X['Outlook'], X['Temp'], c=labels, cmap='viridis', label='Clusters')
plt.title('DBSCAN Clustering')
plt.xlabel('Outlook')
plt.ylabel('Temp')
plt.legend()
plt.show()
```
